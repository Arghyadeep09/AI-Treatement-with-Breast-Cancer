# Explore Dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'breast_cancer_data.csv'
data = pd.read_csv(file_path)
```


# Apply EDA

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Distribution of the target variable
sns.countplot(x='diagnosis', data=data)
plt.title('Distribution of Diagnosis')
plt.show()
```

```python
# Correlation matrix
from sklearn.preprocessing import LabelEncoder

# Encode the target variable (diagnosis) to binary
label_encoder = LabelEncoder()
data['diagnosis'] = label_encoder.fit_transform(data['diagnosis'])

corr_matrix = data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

```

# Data Cleaning

```python
# Drop missing values
data = data.dropna()

# Remove duplicates
data = data.drop_duplicates()

#  Fill missing values with the median (or mean) of the column
data = data.fillna(data.median())

# Check for duplicate rows
duplicates = data.duplicated().sum()
print(f'Number of duplicate rows: {duplicates}')

# Remove duplicate rows
data = data.drop_duplicates()
'''
# Handle outliers using IQR
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
data = data[~((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)]
'''
# Drop the 'id' column as it's not needed
data = data.drop(columns=['id'])
```

```python
data.head()
```

#  Feature Engineering

```python
from sklearn.preprocessing import StandardScaler

# Log transformation on skewed features
skewed_features = data.skew().sort_values(ascending=False)
for feature in skewed_features.index:
    if skewed_features[feature] > 0.5:
        data[feature] = np.log1p(data[feature])

```

# Feature Selection

```python
# Encode the target variable (diagnosis) to binary
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data['diagnosis'] = label_encoder.fit_transform(data['diagnosis'])

# Split into features and labels
X = data.drop('diagnosis', axis=1)
y = data['diagnosis']

# Train a Random Forest model to determine feature importance
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42)
rf.fit(X, y)

# Get feature importances and select the top 5 features
importances = rf.feature_importances_
feature_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

# Select the top 5 features
top_5_features = feature_importance_df['Feature'].head(5).tolist()
print(f'Top 5 Features: {top_5_features}')

# Reduce the dataset to only these top 5 features
X_top_5 = X[top_5_features]
```

```python
from sklearn.preprocessing import StandardScaler

# Scaling features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X_top_5)
scaled_data = pd.DataFrame(scaled_data, columns=X_top_5.columns)

import pickle

# Save the scaler to a file
with open('scaler.pkl', 'wb') as scaler_file:
    pickle.dump(scaler, scaler_file)
```

# Split X, Y

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=42)
print(X_train.columns)
```

# Model Training


```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
from sklearn.linear_model import LogisticRegression


# Initialize model
log_reg = LogisticRegression(max_iter=10000, random_state=42)
log_reg.fit(X_train, y_train)

```

```python
# Make predictions
y_pred = log_reg.predict(X_test)
y_prob = log_reg.predict_proba(X_test)[:,1]

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_prob)

# Print the evaluation metrics
print(f'Accuracy: {accuracy:.2f}')
print('Confusion Matrix:\n', conf_matrix)
print('Classification Report:\n', class_report)
print(f'ROC AUC Score: {roc_auc:.2f}')
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# Load the trained model and scaler
with open('scaler.pkl', 'rb') as scaler_file:
    scaler = pickle.load(scaler_file)

with open('model.pkl', 'rb') as model_file:
    model = pickle.load(model_file)

# Top 5 features from user input (Example values, replace with actual user input)
user_input = {
    'area_worst': 1000.0,
    'concave points_worst': 0.2,
    'radius_worst': 17.5,
    'perimeter_worst': 115.0,
    'area_mean': 700.0
}

# Convert user input to the required format
features = np.array([[
    user_input['area_worst'],
    user_input['concave points_worst'],
    user_input['radius_worst'],
    user_input['perimeter_worst'],
    user_input['area_mean']
]])

# Scale the features
scaled_features = scaler.transform(features)

# Predict probabilities
probabilities = model.predict_proba(scaled_features)[0]

# Display the results
benign_percentage = probabilities[0] * 100
malignant_percentage = probabilities[1] * 100

print(f"Benign: {benign_percentage:.2f}%")
print(f"Malignant: {malignant_percentage:.2f}%")

# Plot pie chart
labels = ['Benign', 'Malignant']
sizes = [benign_percentage, malignant_percentage]
colors = ['lightblue', 'lightcoral']
explode = (0, 0.1)  # explode the "Malignant" slice

plt.figure(figsize=(7, 7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Breast Cancer Prediction')
plt.show()
```
